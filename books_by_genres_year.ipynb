```python
import pandas as pd
```

```python
bestseller_books = "./data/bestsellers_with_categories.csv"
```

```python
bestseller_books_df = pd.read_csv(bestseller_books,encoding="ISO-8859-1")
```

```python
bestseller_books_df.head()
```

```python
only_non_fiction = bestseller_books_df.loc[bestseller_books_df["Genre"] == "Non Fiction", :]

sorted_only_non_fiction = only_non_fiction.sort_values(["Year"], ascending = [True]) 
sorted_only_non_fiction.head()
```

```python
grouped_nonfiction_df = sorted_only_non_fiction.groupby(['Year'])
#print(grouped_nonfiction_df)
```

```python
non_fiction_by_year = pd.DataFrame(grouped_nonfiction_df["Name"].count()).rename(columns={"Name":"non_fiction_counts"})
non_fiction_by_year.head()
```

```python
only_fiction = bestseller_books_df.loc[bestseller_books_df["Genre"] == "Fiction", :]

sorted_only_fiction = only_fiction.sort_values(["Year"], ascending = [True]) 
sorted_only_fiction.head(5)
```

```python
grouped_fiction_df = sorted_only_fiction.groupby(['Year'])
#print(grouped_fiction_df)
```

```python
fiction_by_year = pd.DataFrame(grouped_fiction_df["Name"].count()).rename(columns={"Name":"fiction_counts"})
fiction_by_year.head()
```

```python
genre_by_year_df = pd.concat([fiction_by_year, non_fiction_by_year], axis=1)
genre_by_year_df
```

```python
genre_by_year_df.to_csv("./data/books_by_genres_by_year.csv",encoding="utf-8", index=True, header=True)
```

```python
#books_by_genre_by_year = bestseller_books_df.groupby(['Year','Genre'])
books_by_genre = bestseller_books_df.groupby(['Genre'])
books_by_genre
```

```python
books_by_genre_df = pd.DataFrame(books_by_genre["Name"].count()).rename(columns={"Name":"Counts"})
books_by_genre_df.head(10)
```

```python
books_by_genre_df.to_csv("./data/books_by_genres.csv",encoding="utf-8", index=True, header=True)
```
